```python
import cv2
cap= cv2.VideoCapture(0)
#print(cap.isOpened())
while (cap.isOpened()):
    ret, frame= cap.read()
    width=cap.get(cv2.CAP_PROP_FRAME_WIDTH)
    height=cap.get(cv2.CAP_PROP_FRAME_HEIGHT)
    print(width)
    print(height)
    gray=cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    cv2.imshow('frame', gray)
    if cv2.waitKey(100) & 0xFF== 27:
        break
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
cap= cv2.VideoCapture(0)

fourcc= cv2.VideoWriter_fourcc(*'XVID')#('X','V',"i",'D')
video= cv2.VideoWriter('video_1.mkv', fourcc,30.0, (640, 480) )
#print(cap.isOpened())
while (cap.isOpened()):
    ret, frame= cap.read()
    if ret== True:
        width=cap.get(cv2.CAP_PROP_FRAME_WIDTH)
        height=cap.get(cv2.CAP_PROP_FRAME_HEIGHT)
        print(width)
        print(height)
        video.write(frame)
        gray=cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        cv2.imshow('frame', gray)
        if cv2.waitKey(100) & 0xFF== 27:# or use ord('q')
            break
            
    else:
        break
        
cap.release()
video.release()
cv2.destroyAllWindows()
```
